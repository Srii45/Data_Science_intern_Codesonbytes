```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load train and test datasets from CSV files
train_data = pd.read_csv("C:\\Users\\Srinivas\\Desktop\\Codesonbyte\\test_data.csv")
test_data = pd.read_csv("C:\\Users\\Srinivas\\Desktop\\Codesonbyte\\test_data.csv")

# Separate features and target variables for train and test datasets
X_train, y_train = train_data[['x']], train_data['y']
X_test, y_test = test_data[['x']], test_data['y']
# Create a Linear Regression model
model = LinearRegression()

# Train the model on the train dataset
model.fit(X_train, y_train)

# Make predictions on the test dataset
y_pred = model.predict(X_test)

# Evaluate the model on the test dataset
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE) on test data: {mse}")
print(f"R-squared (R2) on test data: {r2}")


# Visualizations
# Heatmap
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Regression plot
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('X_test')
plt.ylabel('y_test')
plt.title('Linear Regression Fit')
plt.show()

```
